```python
def conversion(c):
  return c * 9/5 + 32

conversion(30)
```

```python
import numpy as np
import pandas as pd

#La experiencia pasada

#caracteristicas de entrada
X = np.array([-30, -10, 0 ,5, 15, 30, 40, 50, 60, 70, 80])
y = np.array([ conversion(c) for c in X ]) # List comprehention

df_temp = pd.DataFrame({'celsius': X, 'fahrenheit': y})
df_temp
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.scatterplot(data=df_temp, x='celsius', y='fahrenheit', hue=df_temp['fahrenheit'], palette='coolwarm')
plt.show()

```

```python
#Regresion Lineal

#[-30, -10,   0,   5,  15,  30,  40,  50,  60,  70,  80]
#[[-30], [-10],   [0],   5,  15,  30,  40,  50,  60,  70,  80]

X_procesadas = df_temp.celsius.values.reshape(-1, 1)
y_procesadas = df_temp.fahrenheit.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression

#Instanciación del modelo
modelo = LinearRegression()

#           caracteristicas, etiquetas
modelo.fit(X_procesadas, y_procesadas)

```

```python
resultado = modelo.predict([[100]])

resultado
```

```python
modelo.score(X_procesadas, y_procesadas)
```

```python
print(modelo.coef_)
print(modelo.intercept_)
```
