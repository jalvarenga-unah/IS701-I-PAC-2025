```python
# df[ df['team_name'] == 'Barcelona'  ][['player_name','goals']]
columnas = ['player_name','goals','assists','xG']
 #jugadores del team y al menos un gol

condicion = (df['team_name'] == 'Barcelona') & (df['goals'] > 0)

estadistica_barsa = df.loc[condicion ,columnas].sort_values(by='player_name')
```

```python
nueva_col = estadistica_barsa['goals'] + estadistica_barsa['assists']

estadistica_barsa['goals_assists'] = nueva_col
estadistica_barsa
```

```python
estadistica_barsa['ratio_goals'] = estadistica_barsa['xG']/ estadistica_barsa['goals']
estadistica_barsa
```

```python
estadistica_barsa = estadistica_barsa.groupby('player_name')[['goals','assists','goals_assists']].sum().reset_index()
```

##Practica

- el jugador con mas tarjetas rojas de cada equipo

- el jugador con menos tarjetas rojas y amarillas de cada equipo

```python
columnas = [ 'player_name','team_name','red_cards','yellow_cards']

# sumary_data = df.groupby(['team_name','player_name'])[['red_cards','yellow_cards']].sum().reset_index()
sumary_data = df.groupby(['team_name','player_name']).agg({
    'red_cards':'sum',
    'yellow_cards':'sum'
}) .reset_index()


# encuentra el valor maximo, pero me devuelve el indice del registro encontrado
max_red_card = sumary_data.loc[sumary_data.groupby('team_name')['red_cards'].idxmax(), columnas]
print('Jugadores con mas tarjetas rojas')
print(max_red_card.sort_values(by='red_cards', ascending=False))

total_cards = sumary_data['red_cards'] + sumary_data['yellow_cards']
sumary_data['total_cards'] = total_cards

columnas = ['player_name','team_name','total_cards']

min_read_yellow_cards = sumary_data.loc[sumary_data.groupby('team_name')['total_cards'].idxmin(), columnas]

print('Jugadores con menor cantidad de tarjetas')
print( min_read_yellow_cards.sort_values(by='total_cards', ascending=False) )
```
