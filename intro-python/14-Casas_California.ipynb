```python
dummies = pd.get_dummies(datos['ocean_proximity'], dtype=int)


datos = pd.concat([datos, dummies], axis=1)

datos
```

```python
datos['room_ratio'] = datos['total_bedrooms'] / datos['total_rooms']
```

```python
X = datos.drop('median_house_value', axis=1)
y = datos['median_house_value']
```

```python
from  sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()

modelo.fit(X_train, y_train)
```

```python
predicciones = modelo.predict(X_test)

pd.DataFrame({'Predicciones': predicciones, 'Valores reales': y_test})
```

```python
print(modelo.score(X_train, y_train))
print(modelo.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_test, predicciones)

import numpy as np

np.sqrt(rmse)
```
