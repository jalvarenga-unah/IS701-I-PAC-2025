```python
def conversion(c):
  return c * 9/5 + 32

conversion(100)
```

```python
import numpy as np
import pandas as pd

#La experiencia pasada

#caracteristicas de entrada
X = np.array([-30, -20, -10, 0 ,5, 10, 30, 40, 50, 60, 70, 80] , dtype=float)
y = np.array([ conversion(c) for c in X ], dtype=float) # List comprehention

```

```python
# !pip install tensorflow
```

```python
from tensorflow import keras
```

```python
# Analisis de la información
import tensorflow as tf

# Definir las dimensiones de nuestro modelo
#( cuantas capas, cuantas neuronas y las caracteristicas de entrada y salida)

modelo = keras.Sequential([
    # capas
    #units = neuronas
    #input_shape =

    keras.Input(shape=(1,)),# La capa de ntrada

    keras.layers.Dense(units=4),# Capa  oculta
    keras.layers.Dense(units=1) # Capa salida

    # keras.layers.Dense(units=1, input_shape=[1] ),
    # keras.layers.Dense(units=4, ), # Capa de salida
    # keras.layers.Dense(units=2, ), # Capa de salida
    # keras.layers.Dense(units=1, ) # Capa de salida

])

# compilar el modelo ( la funcion de activacion y de perdida )
modelo.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss='mean_squared_error')


```

```python

#entrenar
historial = modelo.fit(X, y, epochs=300, verbose=False)
print("Modelo entrenado!")

```

```python
modelo.predict(np.array([100])) # valor esperado: 212
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
import matplotlib.pyplot as plt

y_pred = modelo.predict(X)
plt.plot(X, y, 'ro', label='Valores reales')
plt.plot(X, y_pred, 'b-', label='Predicciones del modelo')
plt.legend()
plt.show()
```
