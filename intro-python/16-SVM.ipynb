```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

dataCancer = load_breast_cancer()
```

```python
X = dataCancer.data
y = dataCancer.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score

kernels = ['linear','rbf','sigmoid']
gammas = [ 0.1, 0.001, 0.0001, 0.00001 ]


for kernel in kernels:
    for gamma in gammas:
      modelo = SVC(kernel=kernel, gamma=gamma)
      modelo.fit(X_train, y_train)
      predicciones = modelo.predict(X_test)
      print(f"kernel: {kernel}, gamma: {gamma}", accuracy_score(y_test, predicciones))
      print(f"kernel: {kernel}, gamma: {gamma}", precision_score(y_test, predicciones))
```

```python
modelo = SVC(kernel='rbf', gamma=0.0001)
modelo.fit(X_train, y_train)
predicciones = modelo.predict(X_test)
print(f"kernel: {kernel}, gamma: {gamma}", accuracy_score(y_test, predicciones))
print(f"kernel: {kernel}, gamma: {gamma}", precision_score(y_test, predicciones))
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

pd.DataFrame(confusion_matrix(y_test, predicciones), columns=["Pred No", "Pred Si"], index=["Real No", "Real Si"])
```
